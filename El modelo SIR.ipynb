```python
#Visualizacion de soluciones de Ecuaciones diferenciales
#El modelo SIR

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#parametros del modelo
beta,gama = 0.4,0.7
tmax = 365
t = np.linspace(0,tmax,10*tmax)
```

```python
#Definicion del lado derecho del modelo

def modSIR(y,t,beta,gama):
    S,I,R = y
    dSdt = -beta*S*I
    dIdt = beta*S*I-gama*I
    dRdt = gama*I
    return dSdt, dIdt, dRdt
```

```python
#Condiciones iniciales 
S0,I0,R0 = 25,5,0
y0= S0,I0,R0

#
solucion = odeint(modSIR,y0,t,args=(beta,gama))
S,I,R=solucion.T
```

```python
solucion.shape
```

```python
#grafica de las soluciones 
%matplotlib inline
fig= plt.figure()
plt.plot(t,S,'b', label="Susceptibles")
plt.plot(t,I,'r', label="Infecciosos")
plt.plot(t,R,'g', label="Recuperados")
plt.xlabel("Tiempo")
plt.ylabel("Numero")
#plt.ylim(0,20)
plt.xlim(0,20)

legend=fig.legend()
legend.get_frame().set_alpha(0.5)

plt.savefig("ModeloSIR1.pdf")
```

```python
#parametros del modelo

a,m = 0.1, 1.2
tmax = 50
t = np.linspace(0,tmax,10*tmax)
```

```python
#Definicion del lado derecho del modelo

def modQ(y,t,a,m):
    S,X = y
    dSdt = 1-S-(S*m)*1.0*X/(a+S)
    dXdt = (S*m)*1.0*X/(a+S)-X
    return dSdt, dXdt
```

```python
#Condiciones iniciales 
S0,X0 = 5,25
y0= S0,X0

#
solucion = odeint(modQ,y0,t,args=(a,m))
S,X=solucion.T
```

```python
#grafica de las soluciones 
%matplotlib inline
fig= plt.figure()
plt.plot(t,S,'b', label="Recurso")
plt.plot(t,X,'r', label="Consumidor")
plt.xlabel("Tiempo")
plt.ylabel("Numero")
plt.ylim(0,5)
plt.xlim(0,50)

legend=fig.legend()
legend.get_frame().set_alpha(0.5)
```
